```python
import numpy                       
from matplotlib import pyplot                 
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
```

```python
nx = 41 
L = 2.0  
dx = L / (nx - 1)  
nt = 25 
dt = 0.02  
c = 1.0  
x = numpy.linspace(0.0, L, num=nx)
```

```python
u= numpy.ones(nx)
mask = numpy.where(numpy.logical_and(x >= 0.5, x <= 1.0))
u[mask] = 2.0
print(u)
```

```python
pyplot.figure(figsize=(8.0, 8.0))
pyplot.title('Initial conditions')
pyplot.xlabel('x')
pyplot.ylabel('u')
pyplot.grid()
pyplot.plot(x, u, color='C0', linestyle='--', linewidth=2)
pyplot.xlim(0.0, L)
pyplot.ylim(0.0, 2.5);
```
